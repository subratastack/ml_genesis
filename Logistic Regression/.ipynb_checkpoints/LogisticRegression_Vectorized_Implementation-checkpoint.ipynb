```python
# import libraries

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# load data
df = pd.read_csv('data/iris-data.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# removing all null value rows
df = df.dropna(subset=['petal_width_cm'])
df.info()
```

```python
# plot
sns.pairplot(df, hue='class', height=2.5)
```

From the above plot we can observe that there are some abnormality in the class name. Let's explore.

```python
# count the unique class
df['class'].value_counts()
```

Two observation can be made here,
1. There are 5 data point marked as versicolor instead of Iris-versicolor
2. There is 1 data point marked as Iris-setossa instead of Iris-setosa

```python
df['class'].replace(['Iris-setossa', 'versicolor'], ['Iris-setosa', 'Iris-versicolor'], inplace=True)
df['class'].value_counts()
```

## Simple Logistic Regression

Consider only 2 class 'Iris-setosa' and 'Iris-versicolor'. Dropping all other class.

```python
final_df = df[df['class'] != 'Iris-virginica']
final_df.head()
```

### Outlier Check

```python
sns.pairplot(final_df, hue='class', height=3)
```

From the above plot we can see that 'sepal_width' and 'sepal_length' has outliers.

SEPAL LENGTH

```python
final_df.hist(column='sepal_length_cm', bins=20, figsize=(10, 5))
```

5 data points are less than 1. They are considered in m so conveting them to cm

```python
final_df.loc[final_df.sepal_length_cm < 1, ['sepal_length_cm']] = final_df['sepal_length_cm']*100
```

```python
final_df.hist(column='sepal_length_cm', bins=20, figsize=(10, 5))
```

SEPAL WIDTH

```python
final_df.hist(column='sepal_width_cm', bins=20, figsize=(10, 5))
```

```python
final_df=final_df.drop(final_df[(final_df['class'] == 'Iris-setosa') & (final_df['sepal_width_cm']  < 2.5)].index)
```

```python
sns.pairplot(final_df, hue='class', height=3)
```

### Label Encoding

```python
final_df['class'].replace(['Iris-setosa', 'Iris-versicolor'], [1, 0], inplace=True)
```
