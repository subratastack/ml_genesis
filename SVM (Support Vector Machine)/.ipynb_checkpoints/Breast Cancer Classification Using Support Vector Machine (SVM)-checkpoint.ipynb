## Breast Cancer Classification Using Support Vector Machine (SVM)

#### Project Task

In this study, my task is to classify tumors into malignant (cancerous) or benign (non-cancerous) using features obtained from several cell images.

Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image.

```python
# load libraries
import pandas as pd
import numpy as np

import plotly.express as px
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

from sklearn.metrics import confusion_matrix, classification_report
```

```python
# load dataset from sklearn datasets
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

Let's convert the cancer dataset into DataFrame and then view

```python
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
df_cancer.head()
```

### Classification Model: Support Vector Machine (SVM)

We will create 2 variable X and y.
- X = Dependent variables
- y = Target variables

```python
X = df_cancer.drop(['target'], axis=1)
X.head()
```

```python
y = df_cancer['target']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

print('Shape of X_train: ', X_train.shape)
print('Shape of X_test: ', X_test.shape)
print('Shape of y_train: ', y_train.shape)
print('Shape of y_test: ', y_test.shape)
```

#### Now we will create an instance of SVC and train our model

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

Now we use our model to predict on test set

```python
y_pred = svc_model.predict(X_test)
```

#### Let's create a confusion matrix to check our model prediction

```python
cm = np.array(confusion_matrix(y_test, y_pred, labels=[1, 0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'], columns=['predicted_cancer', 'predicted_healthy'])
confusion
```

#### Let's now plot the confusion matrix in a heatmap

```python
sns.heatmap(confusion, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

### Let us now normalize the dataset and see how that effects the performance of the SVM model

X' = ($X$-$X_{min}$) / ($X_{max}$ - $X_{min}$) -> $X_{range}$

```python
X_train_min = X_train.min()
X_train_min
```

```python
X_train_max = X_train.max()
X_train_max
```

```python
X_train_range = (X_train_max - X_train_min)
X_train_range
```

```python
X_train_scaled = (X_train - X_train_min) / (X_train_range)
X_train_scaled.head()
```

```python
X_train_scaled.describe()
```

#### Normalize the test dataset as well

```python
X_test_min = X_test.min()
X_test_max = X_test.max()
X_test_range = X_test_max - X_test_min
X_test_scaled = (X_test - X_test_min) / (X_test_range)
```

```python
X_test_scaled.describe()
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train_scaled, y_train)
```

```python
y_scaled_pred = svc_model.predict(X_test_scaled)
```

```python
cm = np.array(confusion_matrix(y_test, y_scaled_pred, labels=[1, 0]))
confusion = pd.DataFrame(confusion, index=['is_cancer', 'is_healthy'], columns=['predicted_cancer', 'predicted_healthy'])
confusion
```

#### Let's plot the confusion matrix in a heatmap

```python
sns.heatmap(confusion, annot=True)
```

```python
print(classification_report(y_test, y_scaled_pred))
```
