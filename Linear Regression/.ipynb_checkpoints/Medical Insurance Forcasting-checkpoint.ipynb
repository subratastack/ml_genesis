### Medical Insurance Forcasting using Linear Regression

```python
# load libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# load the dataset
dataset = pd.read_csv('data/insurance.csv')
dataset.head()
```

```python
# let's check the description of our dataset
dataset.describe()
```

```python
# check for null data
print(dataset.isna().count())
```

```python
# check unique values in sex column
print(dataset['sex'].unique())
```

```python
# converting sex column from string to numeric
dataset['sex_n'] = np.where(dataset['sex'] == 'female', 0, 1)
dataset.head()
```

```python
# check unique values in smoker column
print(dataset['smoker'].unique())
```

```python
# converting smoker column from string to numeric
dataset['smoker_n'] = np.where(dataset['smoker'] == 'yes', 1, 0)
dataset.head()
```

```python
# let's visualize the pair plot
sns.pairplot(dataset)
```

```python
# make a copy of the dataset
dataset_ = dataset
dataset_.head()
```

```python
minmax_transformer = Pipeline(steps=[('minmax', MinMaxScaler())])

preprocessor = ColumnTransformer(
    remainder='drop',
    transformers=[
        ('mm', minmax_transformer, ['age', 'bmi', 'children', 'charges', 'sex_n', 'smoker_n'])
    ]
)

model = preprocessor.fit(dataset_)
scaled_dataset_ = model.transform(dataset_)
print(scaled_dataset_)
```

```python
# convert scaled_dataset_ to dataframe
scaled_df = pd.DataFrame(scaled_dataset_, columns=['age', 'bmi', 'children', 'charges', 'sex_n', 'smoker_n'])
scaled_df.head()
```

```python
# check the dataset description
scaled_df.describe()
```

```python
# create a correlation matrix
correlationMatrix = scaled_df.corr().round(2)
sns.heatmap(data=correlationMatrix, annot=True)
```

```python
# split training and test data

X = pd.DataFrame(np.c_[scaled_df['age'], scaled_df['bmi'], scaled_df['smoker_n']])
Y = scaled_df['charges']

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=1, test_size=0.3)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# create the model

regressor = LinearRegression()
regressor.fit(X_train, y_train)

pred_y = regressor.predict(X_test)
```

```python
# accuracy of the model
rmse = np.sqrt(mean_squared_error(y_test, pred_y))
r2 = r2_score(y_test, pred_y)

print('Accuracy Score')
print('RMSE: ', rmse)
print('R2 Score: ', r2)
```
