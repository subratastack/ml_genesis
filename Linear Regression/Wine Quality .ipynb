## Wine quality prediction using linear regression
__[Data set link](https://archive.ics.uci.edu/ml/datasets/wine+qualityl)__

```python
# import libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

%matplotlib inline
```

```python
# load the dataset
dataset = pd.read_csv('data/winequality-red.csv',  sep=";")
dataset.head()
```

```python
# check for null values
dataset.isnull().sum()
```

```python
# check the shape of the database
dataset.shape
```

```python
# check the description of the dataset
dataset.describe()
```

#### Exploratory data analysis & visualization

```python
# create a histogram for quality data distribution
sns.set(rc={'figure.figsize': (12, 9)})
sns.histplot(data=dataset['quality'], bins=5, kde=True)
plt.show()
```

```python
# create a correlation matrix to see the relationship between variables
correlationMatrix = dataset.corr().round(2)

sns.heatmap(data=correlationMatrix, annot=True)
```

```python
# let's create scatter plot against quality and other variables

features = ['alcohol', 'sulphates', 'volatile acidity', 'citric acid']
target = dataset['quality']

plt.figure(figsize=(20, 5))

for j, column in enumerate(features):
    plt.subplot(1, len(features), j+1)
    x=dataset[column]
    y=target
    plt.scatter(x, y, marker='o')
    plt.title(column)
    plt.xlabel(column)
    plt.ylabel('Quality')
```

#### Modeling and prediction

```python
# create a train and test data set by splitting the main dataset in 80:20 ratio

X = pd.DataFrame(np.c_[dataset['alcohol'], dataset['sulphates'], dataset['alcohol'], dataset['citric acid']])
Y = dataset['quality']

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=1, test_size=0.2)

# print the shape of the new datasets
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### Training and testing the model

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# model evaluation for train data
predicted_y_train = regressor.predict(X_train)
rmse = (np.sqrt(mean_squared_error(y_train, predicted_y_train)))
r2 = r2_score(y_train, predicted_y_train)

print("Evaluation on train dataset")
print("RMSE: {}".format(rmse))
print("R2: {}".format(r2))

print("-------------------------------------------")

predicted_y_test = regressor.predict(X_test)
rmse = (np.sqrt(mean_squared_error(y_test, predicted_y_test)))
r2 = r2_score(y_test, predicted_y_test)

print("Evaluation on test dataset")
print("RMSE: {}".format(rmse))
print("R2: {}".format(r2))
```
